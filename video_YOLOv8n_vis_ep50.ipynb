```python
import cv2
from ultralytics import YOLO

# 모델 초기화
model = YOLO("YOLOv8x_vis_ep50.pt")

# 비디오 로드
video_path = '/Users/being/Desktop/DJI_0736.MP4'
cap = cv2.VideoCapture(video_path)

# 비디오 작성자 설정
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output_video.MP4', fourcc, 30.0, (640, 480))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # BGR에서 RGB로 변환
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # 이미지 추론
    results = model(frame_rgb)

    # 결과 처리 및 표시
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls = int(box.cls[0])
            label = model.names[cls]
            # 클래스에 따라 다른 색상을 사용하여 바운딩 박스 그리기
            color = (0, 255, 0) if label == 'person' else (0, 0, 255)  # 예시 색상
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

    # 처리된 프레임 저장
    out.write(frame)

    # 프레임 보여주기 (선택적)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 모든 처리가 끝난 후
cap.release()
out.release()
cv2.destroyAllWindows()
```
